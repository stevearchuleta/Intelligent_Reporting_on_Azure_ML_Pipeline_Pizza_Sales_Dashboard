## **Problem Statement**

### **Business Context**

A moderately sized, locally popular food joint operates a chain of outlets located in a metropolitan area, offering a diverse menu of pizzas, sides, and beverages. Despite having a steady flow of customers, they face challenges in optimizing their order fulfillment process, leading to delays during peak hours, which results in customer dissatisfaction and impacts repeat business. Additionally, they struggle with inventory management, often experiencing shortages of popular ingredients or excess stock of less favored items. To address these issues, they are implementing a new order management system and seeking to analyze sales data to better predict demand and streamline inventory.

### **Objective**

The business has engaged Steve Archuleta as a Data Analyst to enhance operational efficiency and boost customer satisfaction. Steve has been provided raw historical sales data and has been tasked with pre-processing historical sales data to uncover trends, building an interactive dashboard to enable visual reporting of key metrics, and generating email reports to communicate key insights to stakeholders. This analysis will enable the stakeholders to get a clearer understanding of the business, stay on top of changing market scenarios via frequent alerts, and make quick, informed decisions to resolve operational challenges. The anticipated outcomes include reduced order processing times, improved inventory turnover, and increased customer satisfaction leading to higher repeat sales.

### **Data Description:**

This dataset contain detailed information about pizza orders, including specifics about the pizza variants, quantities, pricing, dates, times, and categorization details.
- **pizza_id:** A unique identifier assigned to each distinct pizza variant available for ordering.  
- **order_id:** A unique identifier for each order made, which links to multiple pizzas.  
- **pizza_name_id:** An identifier linking to a specific name of the pizza.  
- **quantity:** The number of units of a specific pizza variant ordered within an order.  
- **order_date:** The date when the order was placed.  
- **order_time:** The time when the order was placed.  
- **unit_price:** The cost of a single unit of the specific pizza variant.  
- **pizza_size:** Represents the size of the pizza (e.g., small, medium, large).  
- **pizza_category:** Indicates the category of the pizza, such as vegetarian, non-vegetarian, etc.  
- **pizza_name:** Specifies the name of the specific pizza variant ordered.

### **Connect to Azure Machine Learning Workspace**

```python
# ========================================================================
# IMPORT THE REQUIRED AZURE ML MODULES TO MANAGE THE WORKSPACE AND COMPUTE
# ========================================================================
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential
from azure.ai.ml.entities import AmlCompute

# =====================================================
# CREATE A DEFAULT CREDENTIAL OBJECT FOR AUTHENTICATION
# =====================================================
credential = DefaultAzureCredential()

# ==============================================================
# CREATE A MLCLIENT INSTANCE TO COMMUNICATE WITH AZURE WORKSPACE
# ==============================================================
ml_client = MLClient(
    credential=credential,
    subscription_id="af6e7bea-cbc6-4f83-ab20-3169af7fccdf",
    resource_group_name="default_resource_group",
    workspace_name="pizza-sales-intelligent-reporting"
)

# ==================================
# DEFINE THE NAME OF THE CPU CLUSTER
# ==================================
cpu_compute_target = "cpu-cluster"

# =======================================================
# CHECK IF THE CLUSTER ALREADY EXISTS OR CREATE A NEW ONE
# =======================================================
try:
    # ==== TRY TO GET EXISTING CLUSTER
    cpu_cluster = ml_client.compute.get(cpu_compute_target)
    print(f"ALREADY USING EXISTING CLUSTER NAMED {cpu_compute_target}.")
except Exception:
    print("CREATING A NEW CPU COMPUTE TARGET...")
    # ==== CREATE A NEW CPU CLUSTER WITH DEFINED PARAMETERS
    cpu_cluster = AmlCompute(
        name=cpu_compute_target,
        type="amlcompute",
        size="Standard_A2_v2",
        min_instances=0,
        max_instances=1,
        idle_time_before_scale_down=180,
        tier="Dedicated"
    )
    # ==== CREATE OR UPDATE THE CLUSTER IN AZURE ML WORKSPACE
    cpu_cluster = ml_client.compute.begin_create_or_update(cpu_cluster).result()
    print(f"COMPUTE CLUSTER {cpu_cluster.name} CREATED WITH SIZE {cpu_cluster.size}.")
```

### **Create a Job Environment with Conda YAML**

```python
# =========================================================
# CREATE A DIRECTORY FOR STORING THE CONDA ENVIRONMENT FILE
# =========================================================
src_dir_env = "./project_env"
os.makedirs(src_dir_env, exist_ok=True)

# ==========================================================
# WRITE THE CONDA.YML FILE WITH DEPENDENCIES FOR THE PROJECT
# ==========================================================
conda_yaml_content = """name: sklearn-env
channels:
  - conda-forge
dependencies:
  - python=3.8
  - pip=21.2.4
  - scikit-learn=0.23.2
  - scipy=1.7.1
  - pip:
    - mlflow==2.8.1
    - azureml-mlflow==1.51.0
    - azureml-inference-server-http
    - azureml-core==1.57.0
    - cloudpickle==1.6.0
    - huggingface_hub
    - datasets
    - seaborn
    - pandas
"""

with open(os.path.join(src_dir_env, "conda.yml"), "w") as f:
    f.write(conda_yaml_content)

# =======================================================
# CREATE AN ENVIRONMENT OBJECT USING THE ABOVE CONDA YAML
# =======================================================
from azure.ai.ml.entities import Environment, BuildContext
env_docker_conda = Environment(
    image="mcr.microsoft.com/azureml/openmpi4.1.0-ubuntu20.04",
    conda_file="./project_env/conda.yml",
    name="intelligent-reporting-azure",
    description="ENVIRONMENT WITH CONDA DEPENDENCIES FOR PIZZA SALES ANALYSIS"
)

# =======================================
# CREATE OR UPDATE THE ENVIRONMENT IN AML
# =======================================
ml_client.environments.create_or_update(env_docker_conda)
```

### **Feature Engineering**

```python
# =====================================================
# FEATURE EXTRACTION FOR CATEGORICAL/TEXTUAL DATA
# INGREDIENTS: COUNT FREQUENCY OF INDIVIDUAL INGREDIENTS
# =====================================================

# THE GOAL:
# - DETERMINE THE CORRECT DELIMITER (" ,") FOR INGREDIENTS
# - SPLIT THE 'pizza_ingredients' COLUMN INTO INDIVIDUAL INGREDIENTS
# - CHECK IF THERE ARE LEADING/TRAILING SPACES
# - CLEAN AND STANDARDIZE THE INGREDIENT NAMES
# - COUNT FREQUENCY OF EACH INGREDIENT

# IMPORTANT NOTE:
# INITIAL ATTEMPTS SHOWED THAT SPLITTING BY JUST "," LEFT LEADING SPACES.
# THEREFORE, THE CHOSEN DELIMITER IS ", " (COMMA + SPACE) FOR A CLEANER SEPARATION.
```

```python
# ============================================================
# CREATE A SERIES OF ALL INGREDIENTS FROM ALL PIZZAS
# SPLIT THE 'pizza_ingredients' COLUMN BY ", " (COMMA + SPACE)
# THEN EXPLODE THE LISTS INTO INDIVIDUAL ROWS
# ============================================================
all_ingredients = (
    df['pizza_ingredients']
    .apply(lambda x: x.split(', '))  # split by comma + space for cleaner separation
    .explode()                       # transform lists into separate rows
)

all_ingredients
```

```python
# ========================================================
# CHECK FOR LEADING AND TRAILING SPACES
# CONFIRM THAT SPLITTING BY ", " ELIMINATED SPACING ISSUES
# ========================================================
starts_with_space = all_ingredients.str.startswith(' ').any()
ends_with_space = all_ingredients.str.endswith(' ').any()

print("Any ingredient starts with a space:", starts_with_space)
print("Any ingredient ends with a space:", ends_with_space)
```

```python
# ===========================================================
# ADDITIONAL CLEANING STEPS
# ENSURE CONSISTENT CASING AND REMOVE TRAILING/LEADING SPACES
# (NOT STRICTLY NECESSARY, BUT THIS STEP ENSURES UNIFORMITY)
# ===========================================================
all_ingredients = all_ingredients.str.strip().str.lower()

all_ingredients
```

```python
# ======================================
# COUNT THE FREQUENCY OF EACH INGREDIENT
# ======================================
ingredient_counts = all_ingredients.value_counts()
ingredient_counts
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div> 
<ul> 
<li><div style="color: blue;">THE LIST DISPLAYS INGREDIENTS IN DESCENDING ORDER OF FREQUENCY, INDICATING WHICH INGREDIENTS ARE MOST COMMON.</div></li> 
<li><div style="color: blue;">THE NUMBERS REPRESENT THE TOTAL NUMBER OF TIMES EACH INGREDIENT APPEARS ACROSS ALL PIZZA ORDERS IN THE DATASET.</div></li> 
<li><div style="color: blue;">INGREDIENTS LIKE GARLIC AND TOMATOES HAVE VERY HIGH COUNTS (27,422 AND 26,601 RESPECTIVELY), SUGGESTING THEY ARE STAPLE INGREDIENTS IN MANY PIZZAS.</div></li> 
<li><div style="color: blue;">INGREDIENTS WITH LOWER COUNTS, SUCH AS PEARS OR THYME (480 OCCURRENCES), ARE RARER, INDICATING THEY ARE USED LESS FREQUENTLY.</div></li> 
</ul>

```python
# =================================================================
# //COMBINE order_date AND order_time INTO A SINGLE DATETIME COLUMN
# //DROP THE ORIGINAL order_date COLUMN
# //DISPLAY THE FIRST FEW ROWS OF THE MODIFIED df DATAFRAME
# =================================================================
df['order_time'] = df['order_date'].combine(
    df['order_time'], 
    lambda d, t: d.replace(hour=t.hour, minute=t.minute, second=t.second)
)
df.drop(columns=['order_date'], inplace=True)
df.head()
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div> 
<ul> 
<li><div style="color: blue;">THE order_date COLUMN HAS BEEN REMOVED, AND A UNIFIED order_time COLUMN NOW REPRESENTS BOTH DATE AND TIME.</div></li> 
<li><div style="color: blue;">THE MODIFIED DATAFRAME SHOWS THAT ORDER TIMING IS PRECISELY CAPTURED WITH HOURS, MINUTES, AND SECONDS.</div></li> 
<li><div style="color: blue;">ALL OTHER ORIGINAL FEATURES REMAIN, ENSURING CONTINUITY OF THE DATA WHILE ENHANCING TEMPORAL ANALYSIS CAPABILITY.</div></li> 
</ul>


```python
# ====================================================
# //EXTRACT MONTH FROM order_time INTO order_month
# //CLASSIFY order_time INTO time_of_day BASED ON HOUR:
# //     MORNING = 6 < HOUR < 12
# //     AFTERNOON = 12 < HOUR < 18
# //     EVENING = OTHERWISE
# //DISPLAY FIRST FEW ROWS OF MODIFIED DATAFRAME
# ====================================================
df['order_month'] = df['order_time'].dt.month

df['time_of_day'] = df['order_time'].dt.hour.apply(
    lambda x: "Morning" if 6 < x < 12 
    else ("Afternoon" if 12 < x < 18 
    else "Evening")
)

df.head()
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div> 
<ul> 
<li><div style="color: blue;">A NEW order_month COLUMN SUCCESSFULLY EXTRACTS THE MONTH AS A NUMERIC VALUE, FACILITATING MONTHLY TREND ANALYSIS.</div></li> 
<li><div style="color: blue;">A NEW time_of_day COLUMN CLASSIFIES ORDERS INTO MORNING, AFTERNOON, OR EVENING, ADDING A USEFUL CATEGORICAL TIME FEATURE.</div></li> 
<li><div style="color: blue;">THE DATAFRAME NOW PROVIDES MORE GRANULAR TIME-BASED INSIGHTS WITHOUT ALTERING OTHER EXISTING FEATURES.</div></li>

```python
# ==============================================
# VIEW DESCRIPTIVE STATISTICS FOR total_price
# IDENTIFY POTENTIAL RANGE AND SCALE OF OUTLIERS
# ==============================================
df['total_price'].describe()
```

```python
# =========================================================
# IDENTIFY OUTLIERS USING THE INTERQUARTILE RANGE (IQR)
# POINTS BEYOND 1.5 * IQR ABOVE Q3 OR BELOW Q1 ARE OUTLIERS
# =========================================================
Q1 = df['total_price'].quantile(0.25)
Q3 = df['total_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print('\n')
print("IQR:", IQR)
print('\n')
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print('\n')
# COUNT OUTLIERS
outliers_lower = (df['total_price'] < lower_bound).sum()
outliers_upper = (df['total_price'] > upper_bound).sum()

print("Number of Outliers Below Lower Bound:", outliers_lower)
print("Number of Outliers Above Upper Bound:", outliers_upper)
```

```python
# ==============================================
# //INSPECT EXTREME VALUES ABOVE THE UPPER BOUND
# //MAY PROVIDE CLUES ABOUT WHY THEY OCCUR
# ==============================================
extreme_values = df[df['total_price'] > upper_bound]
extreme_values.head(10)  # SHOW THE TOP 10 EXTREME ROWS
```

```python
# ==========================================================
# HANDLE OUTLIERS (BASED ON RESULTS FROM ABOVE
# UPPER_BOUND = 32.125 (IDENTIFIED FROM IQR METHOD)
# OUTLIERS SEEM TO REPRESENT MULTI-PIZZA ORDERS, NOT ERRORS!
# ==========================================================

# ORIGINAL DATAFRAME FOR REFERENCE
# df_original = df.copy()

# STRATEGY 1: CAP AT UPPER BOUND
# NOTE: THIS KEEPS THE DATA WITHIN A REDUCED RANGE BUT MAY DISTORT LEGITIMATE LARGE ORDERS
# df_capped = df.copy()
# df_capped.loc[df_capped['total_price'] > 32.125, 'total_price'] = 32.125

# STRATEGY 2: REMOVE OUTLIERS ENTIRELY
# NOTE: THIS ELIMINATES LEGITIMATE HIGH-VALUE ORDERS, REDUCING SAMPLE SIZE AND POTENTIALLY BIASING ANALYSIS
# df_no_outliers = df[df['total_price'] <= 32.125]

# STRATEGY 3: APPLY A TRANSFORMATION (E.G., LOG TRANSFORM)
# NOTE: LOG TRANSFORMATION CAN REDUCE THE IMPACT OF EXTREME VALUES WITHOUT DISCARDING OR CAPPING THEM
df_log_transformed = df.copy()
df_log_transformed['total_price'] = np.log1p(df_log_transformed['total_price'])

# USING THE LOG-TRANSFORMED DATAFRAME TO REDRAW THE BOXPLOT
```

```python
# ==========================================================
# PRINCIPAL COMPONENT ANALYSIS (PCA) FOR DIMENSION REDUCTION
# REDUCE NUMERIC FEATURES TO TWO PRINCIPAL COMPONENTS
# PLOT THE FIRST TWO COMPONENTS WITH HUE AS pizza_category
# ==========================================================
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

numeric_features = df[['quantity', 'unit_price', 'total_price']].dropna()
scaled_data = StandardScaler().fit_transform(numeric_features)

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_components, columns=['PC1', 'PC2'])
pca_df['pizza_category'] = df.loc[numeric_features.index, 'pizza_category']

plt.figure(figsize=(10,6))
ax = sns.scatterplot(x='PC1', y='PC2', hue='pizza_category', data=pca_df, palette='Set2', alpha=0.7)

plt.title("PCA SCATTER PLOT: REDUCED DIMENSIONS WITH CATEGORY HUE")
plt.xlabel("PRINCIPAL COMPONENT 1")
plt.ylabel("PRINCIPAL COMPONENT 2")
plt.grid(True, linestyle='--')

# Add annotation explaining PCA in the lower right quadrant
annotation_text = (
    "PCA (Principal Component Analysis) reduces high-dimensional data\n"
    "into fewer components (PC1, PC2) capturing the most variance.\n"
    "• PC1 & PC2 are new axes that explain the largest trends in data.\n"
    "• This scatterplot shows data spread in just two dimensions,\n"
    "  making patterns easier to visualize.\n"
    "• By coloring points by category, we can see if certain categories\n"
    "  cluster together or separate out along these principal axes."
)

# Get current axis limits
x_min, x_max = ax.get_xlim()
y_min, y_max = ax.get_ylim()

# Place annotation in lower right quadrant
ax.text(
    x_max*0.3, y_min*0.6,   # Adjust these multipliers as needed to position text
    annotation_text,
    fontsize=10,
    color="black",
    va="bottom",
    ha="left",
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", alpha=0.8)
)

plt.show()
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div> 
<ul> 
<li><div style="color: blue;">AFTER REDUCING THE DATA TO TWO PRINCIPAL COMPONENTS (PC1 AND PC2), EACH POINT REPRESENTS A PIZZA ORDER THAT HAS BEEN PLOTTED ACCORDING TO THE MOST SIGNIFICANT PATTERNS CAPTURED IN THE NUMERIC FEATURES (quantity, unit_price, total_price).</div></li> 
<li><div style="color: blue;">EVEN THOUGH THE ORIGINAL DATA WAS MULTIDIMENSIONAL, THE PLOT SHOWS THAT CERTAIN PIZZA CATEGORIES TEND TO GROUP TOGETHER ALONG THESE NEW AXES. FOR EXAMPLE, IF Classic and Veggie TEND TO CLUSTER IN A SIMILAR AREA, IT SUGGESTS THEIR ORDERS SHARE COMMON TRAITS IN PRICE AND QUANTITY PATTERNS.</div></li> 
<li><div style="color: blue;">PC1 AND PC2 DON’T REPRESENT DIRECT FACTORS LIKE PRICE OR SIZE, BUT RATHER NEW DIMENSIONS THAT COMBINE THESE FEATURES TO EXPLAIN THE MOST VARIANCE IN THE DATASET. ORDERS THAT APPEAR CLOSER TOGETHER LIKELY HAVE SIMILAR SPENDING LEVELS AND PRODUCT MIXES, WHILE THOSE FAR APART MAY REFLECT DISTINCT ORDERING BEHAVIORS.</div></li> 
<li><div style="color: blue;">COLORING THE POINTS BY pizza_category HELPS DETERMINE IF CERTAIN CATEGORIES GRAVITATE TOWARDS PARTICULAR REGIONS IN THIS PCA SPACE. IF A SPECIFIC CATEGORY CONSISTENTLY PLOTS HIGHER ON PC2 OR FARTHER ALONG PC1, THAT SUGGESTS ITS ORDERS MIGHT DIFFER IN KEY WAYS FROM OTHER CATEGORIES (E.G., TYPICALLY MORE EXPENSIVE OR TYPICALLY LARGER IN QUANTITY).</div></li> 
</ul>


```python
# =====================================================================
# RADAR/POLAR PLOT (PARALLEL CATEGORIES) FOR MULTIVARIATE VIEW
# COMPARE MEAN FEATURES ACROSS MULTIPLE pizza_categories
# AGGREGATE MEAN OF quantity, unit_price, total_price FOR EACH CATEGORY
# THEN NORMALIZE AND PLOT ON A RADAR CHART
# =====================================================================


# ==== CALCULATE MEAN OF SELECTED NUMERIC FEATURES BY pizza_category
radar_data = df.groupby('pizza_category')[['quantity','unit_price','total_price']].mean()

# ==== NORMALIZE THE VALUES TO RANGE [0,1] FOR RADAR PLOT
radar_data_norm = (radar_data - radar_data.min()) / (radar_data.max() - radar_data.min())
categories = radar_data_norm.columns.tolist()
N = len(categories)

# ==== RADAR CHART SETUP
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(10,8))
ax = plt.subplot(111, polar=True)

for cat in radar_data_norm.index:
    values = radar_data_norm.loc[cat].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, label=cat, linewidth=2)
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_yticklabels([])
ax.set_title("RADAR PLOT: COMPARING FEATURES ACROSS CATEGORIES (MULTIVARIATE)")
ax.grid(True, linestyle='--')
plt.legend(bbox_to_anchor=(0.6,0.7))
plt.show()
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div> 
<ul> 
<li><div style="color: blue;">A RADAR PLOT IS A WAY TO VISUALIZE MULTIPLE VARIABLES FOR DIFFERENT CATEGORIES ON A SINGLE CHART. EACH AXIS REPRESENTS ONE FEATURE, AND ALL AXES RADIATE OUT FROM THE CENTER, ALLOWING DIRECT COMPARISONS OF CATEGORY "SHAPES" ACROSS MULTIPLE DIMENSIONS.</div></li> 
<li><div style="color: blue;">NORMALIZING THE MEAN VALUES (SCALING THEM BETWEEN 0 AND 1) ENSURES THAT EACH FEATURE IS MEASURED ON THE SAME RELATIVE SCALE. THIS PREVENTS ANY SINGLE FEATURE WITH INHERENTLY LARGER VALUES FROM DOMINATING THE PLOT, MAKING IT EASIER TO SEE RELATIVE DIFFERENCES BETWEEN CATEGORIES.</div></li> 
<li><div style="color: blue;">IN THIS RADAR PLOT, EACH PIZZA CATEGORY IS REPRESENTED BY A POLYGON FORMED BY CONNECTING POINTS ON THE AXES FOR quantity, unit_price, AND total_price. OBSERVING THE SHAPE AND SIZE OF EACH POLYGON HELPS IDENTIFY WHICH CATEGORIES TEND TO HAVE HIGHER OR LOWER NORMALIZED VALUES FOR THESE FEATURES.</div></li> 
<li><div style="color: blue;">FOR EXAMPLE, IF THE POLYGON FOR THE "CHICKEN" CATEGORY EXTENDS FURTHER ALONG THE unit_price AXIS, IT INDICATES THAT CHICKEN PIZZAS HAVE RELATIVELY HIGHER UNIT PRICES COMPARED TO OTHER CATEGORIES. SIMILARLY, A CATEGORY'S POLYGON THAT IS SMALLER OR CLOSER TO THE CENTER ON ALL AXES SUGGESTS THAT IT GENERALLY RANKS LOWER ACROSS THESE NORMALIZED MEAN FEATURES.</div></li> 
<li><div style="color: blue;">BY COMPARING THESE POLYGON SHAPES, VIEWERS CAN QUICKLY SEE PATTERNS: MAYBE ONE CATEGORY CONSISTENTLY SCORES HIGHER ON total_price, WHILE ANOTHER BALANCES OUT BETWEEN quantity AND unit_price. THE RADAR CHART’S ABILITY TO DISPLAY MULTIVARIATE DATA IN A SIMPLE, COMPARATIVE MANNER HELPS IDENTIFY WHICH CATEGORIES STAND OUT, WHICH ARE SIMILAR, AND WHERE POTENTIAL OPPORTUNITIES OR ANOMALIES LIE.</div></li>

## **Data Preprocessing**


- ### Create Preprocessing Script

- Handle duplicates, missing values, date/time processing, and feature engineering for order_time, order_month, and time_of_day.

```python
# ====================================================
# CREATE A DIRECTORY TO STORE THE PREPROCESSING SCRIPT
# ====================================================
src_dir_job_scripts = "./project_scripts"
os.makedirs(src_dir_job_scripts, exist_ok=True)

# ===============================================================
# WRITE THE PRE_PROCESS.PY SCRIPT TO PROCESS THE PIZZA SALES DATA
# ===============================================================
preprocess_script_content = r'''
# =========================
# IMPORT REQUIRED LIBRARIES
# =========================
import argparse
import pandas as pd
from pathlib import Path
import numpy as np

# =============================================
# DEFINE A FUNCTION TO MAP HOURS TO TIME OF DAY
# =============================================
def map_time_of_day(x):
    # ==== IF HOUR BETWEEN 6 AND 12 => MORNING
    # ==== IF HOUR BETWEEN 12 AND 18 => AFTERNOON
    # ==== OTHERWISE => EVENING
    if 6 < x < 12:
        return "Morning"
    elif 12 < x < 18:
        return "Afternoon"
    else:
        return "Evening"

if __name__ == "__main__":
    # ==========================================
    # PARSE ARGUMENTS FOR INPUT AND OUTPUT PATHS
    # ==========================================
    parser = argparse.ArgumentParser()
    parser.add_argument("--data", type=str, help="PATH TO INPUT DATA")
    parser.add_argument("--output", type=str, help="PATH TO OUTPUT DATA")
    args = parser.parse_args()

    # ===========================================
    # READ THE INPUT DATA FROM THE SPECIFIED PATH
    # ===========================================
    df = pd.read_csv(args.data)

    # =================
    # REMOVE DUPLICATES
    # =================
    df = df.drop_duplicates()

    # ==========================================================
    # HANDLE MISSING VALUES BY DROPPING ROWS WITH MISSING VALUES
    # (ADJUST AS NEEDED FOR BUSINESS LOGIC)
    # ==========================================================
    df = df.dropna()

    # ==============================================================
    # CREATE A SINGLE DATETIME COLUMN FROM order_date AND order_time
    # 'format="mixed"' IS A PLACEHOLDER; USE APPROPRIATE PARSING LOGIC
    # ==============================================================
    df['order_time'] = pd.to_datetime(df['order_date'] + " " + df['order_time'], errors='coerce')

    # ========================================================
    # DROP THE order_date COLUMN AFTER MERGING INTO order_time
    # ========================================================
    df.drop(columns=['order_date'], inplace=True)

    # ==================================================
    # EXTRACT THE MONTH FROM order_time INTO order_month
    # ==================================================
    df['order_month'] = df['order_time'].dt.month

    # ====================================================
    # MAP THE HOUR OF order_time TO A TIME_OF_DAY CATEGORY
    # ====================================================
    df['time_of_day'] = df['order_time'].dt.hour.apply(map_time_of_day)

    # ===============================================================
    # //ONE-HOT ENCODE THE pizza_category COLUMN FOR BETTER MODEL INPUT
    # //AVOID ARTIFICIAL ORDERING ON CATEGORIES
    # //CREATE NEW BINARY INDICATOR COLUMNS
    # ===============================================================
    # Instead of overwriting df directly with get_dummies:
    temp_dummies = pd.get_dummies(df["pizza_category"], prefix="category", drop_first=False)
    df = pd.concat([df, temp_dummies], axis=1)
    # NOTE: do not drop df["pizza_category"] or daily_email_report.py will fail
    # df = pd.get_dummies(df, columns=['pizza_category'], prefix='category', drop_first=False)

    # ===================================================
    # DEFINE THE OUTPUT PATHS AND SAVE THE PROCESSED DATA
    # ===================================================
    output_data_store = args.output
    preprocessed_data_output_path = Path(output_data_store, "processed_data.csv")
    df.to_csv(preprocessed_data_output_path, index=False)
'''

with open(os.path.join(src_dir_job_scripts, "pre_process.py"), "w") as f:
    f.write(preprocess_script_content)
```

<div style="color: orange; font-weight:bold; font-size:large;">👆 OBSERVATION</div>  
<ul>  
<li><div style="color: blue;">THIS CODE BLOCK CREATES A DIRECTORY AND WRITES A PREPROCESSING SCRIPT NAMED pre_process.py. THE SCRIPT EXECUTES DATA CLEANING, DERIVES NEW FEATURES LIKE order_month AND time_of_day, AND ENSURES THAT RAW PIZZA SALES DATA IS PREPARED FOR SUBSEQUENT STEPS IN THE MACHINE LEARNING PIPELINE.</div></li>  
<li><div style="color: blue;">ALTHOUGH NO DIRECT OUTPUT APPEARS HERE, THE EXECUTION OF THESE STEPS FORMS A FOUNDATION FOR A REPRODUCIBLE, RELIABLE ANALYTICAL WORKFLOW, IMPROVING OVERALL DATA PROCESSING CONSISTENCY.</div></li>  
<li><div style="color: blue;">THE ADDITION OF ONE-HOT ENCODING FOR THE pizza_category FEATURE CREATES MULTIPLE BINARY INDICATOR COLUMNS, ONE FOR EACH CATEGORY. THIS APPROACH ENSURES THAT NO ARTIFICIAL ORDERING IS IMPOSED ON CATEGORIES, FACILITATING MORE ACCURATE MODELING AND ANALYSIS BY TREATING EACH CATEGORY AS A DISTINCT, INDEPENDENT FEATURE.</div></li>  
<li><div style="color: blue;">THE COMBINATION OF DATA CLEANING, FEATURE ENGINEERING, AND ONE-HOT ENCODING PRODUCES A HIGH-QUALITY, STRUCTURED DATASET SUITABLE FOR TRAINING MODELS AND CONDUCTING DETAILED ANALYSES WITHOUT INTRODUCING BIASES OR MISLEADING NUMERIC RELATIONSHIPS BETWEEN CATEGORIES.</div></li>  
<li><div style="color: blue;">THE RESULT IS AN AUTOMATED, PORTABLE, AND SCALABLE PREPROCESSING STEP THAT CAN BE REUSED AS THE PIPELINE EVOLVES, ENSURING THAT FUTURE WORKLOADS RECEIVE THE SAME WELL-PREPARED DATA FOUNDATION.</div></li>

        - #A. Data preparation step (processes the input CSV dataset, pizza_sales.csv).

```python
# ==============================================
# IMPORT NECESSARY MODULES FOR PIPELINE CREATION
# ==============================================
# from azure.ai.ml import dsl

# ============================
# DEFINE THE PIPELINE FUNCTION
# ============================
@dsl.pipeline(
    compute='cpu-cluster',
    description="PROJECT - INTELLIGENT REPORTING PIPELINE FOR PIZZA SALES"
)

def project_pipeline_(pipeline_job_data_input, hugging_face_folders):
    
    #A. ==== JOB 1: DATA PREPARATION JOB
    data_prep_job = step_process_pre_process(data=pipeline_job_data_input)

    #B. ==== JOB 2: HUGGING FACE AUTHENTICATION AND FILE PUSH JOB
    hugging_face_job = step_hugging_auth(
        processed_data_push=data_prep_job.outputs.processed_data,
        streamlit_files=hugging_face_folders
    )

    #C. ==== JOB 3: EMAIL REPORTING JOB
    # event_job is email_job... consider changing this variable name
    event_job = step_process_report(
        processed_data=data_prep_job.outputs.processed_data
    )

    # ==== RETURN PIPELINE OUTPUTS
    return {"pipeline_job_prep_data": data_prep_job.outputs.processed_data}

# ==============================================================
# DEFINE THE PATHS FOR INPUT REGISTERED DATA AND STREAMLIT FILES
# ==============================================================
datastore_data_asset = ml_client.data.get("pizza-analysis-data", version="1").path
streamlit_files_path = "./project_files"

# ==================================
# CREATE AN INSTANCE OF THE PIPELINE
# ==================================
pipeline = project_pipeline_(
    pipeline_job_data_input=Input(type="uri_file", path=datastore_data_asset),
    hugging_face_folders=Input(type="uri_folder", path=streamlit_files_path)
)

# =======================
# SUBMIT THE PIPELINE JOB
# =======================
pipeline_job = ml_client.jobs.create_or_update(
    pipeline,
    experiment_name="Project_01_Pizza_Sales_Intelligent_Reporting"
)
```
